# L1范数等距线
《线性代数》 | 鸢尾花书：数学不难

这段代码从数学角度展示了 **L1 范数**（曼哈顿范数）的几何形状，并将其与标准基向量 $e_1 = (1, 0)$ 和 $e_2 = (0, 1)$ 一起在二维平面中可视化。

---

首先，L1 范数定义为向量 $\mathbf{x} = (x_1, x_2)$ 的绝对值和，即：

$$
\|\mathbf{x}\|_1 = |x_1| + |x_2|
$$

代码中的变量 `xx` 和 `yy` 构成了一个二维网格，覆盖了从 $[-8, 8] \times [-8, 8]$ 的平面区域。这相当于在平面上生成了 $1000 \times 1000$ 个点：

$$
(xx, yy) = \{(x, y) \mid x, y \in [-8, 8] \}
$$

接着，对网格上的每个点 $(x, y)$ 计算其对应的 L1 范数值：

$$
zz = |x| + |y|
$$

得到的是一个二维数组 `zz`，其中每个元素代表对应网格点的 L1 范数。

---

在几何上，**L1 范数的等距线（即 $\|\mathbf{x}\|_1 = c$ 的集合）是菱形**，其形状由以下等式描述：

$$
|x_1| + |x_2| = c, \quad c \in \mathbb{R}^+
$$

例如，当 $c = 1$ 时，等距线对应于一个以原点为中心、顶点在 $(1, 0), (0, 1), (-1, 0), (0, -1)$ 的正方形（旋转了45°，看起来像菱形）。

---

代码随后绘制了从 $c = 1$ 到 $c = 8$ 的一系列等高线：

$$
\{|x| + |y| = c \mid c = 1, 2, ..., 8\}
$$

这些等距线层层展开，形成一组菱形轮廓图，这正是 L1 范数等值集合的几何表示。

---

接下来绘制的是二维标准正交基向量：

- $e_1 = (1, 0)$，即 x 轴正方向的单位向量
- $e_2 = (0, 1)$，即 y 轴正方向的单位向量

使用 `plt.quiver` 函数从原点出发分别绘制这两个向量。这两个向量定义了二维空间中最基本的坐标方向。

---

最终图像将包括：

- 黑色的标准基向量箭头；
- 多个颜色渐变的 L1 范数等距线（等高线）；
- 坐标轴、网格线；
- 设置坐标轴范围为 $[-8, 8]$；
- 设置坐标轴比例为 1:1，以保真实几何形状（否则菱形会被拉伸）。

---

这段代码的核心思想是**用可视化方式展示 L1 范数的几何结构及其与标准基之间的关系**，非常适合用于教学“范数”概念、凸集几何结构或优化问题中的可行域展示。

## 初始化

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## 生成数据

```python
xx, yy = np.meshgrid(np.linspace(-8, 8, 1000), 
                     np.linspace(-8, 8, 1000))
```

## 计算 L1 范数

```python
zz = np.abs(xx) + np.abs(yy)  # L1 范数计算公式
```

## 定义标准基向量

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])

# 定义颜色
color_e1 = [0, 0, 0]  # 黑色
color_e2 = [0, 0, 0]  # 黑色
```

## 可视化

```python
plt.figure(figsize=(6, 6))

# 绘制标准基向量 e1 和 e2
plt.quiver(0, 0, e1[0], e1[1], 
           angles='xy', scale_units='xy', 
           scale=1, color=color_e1, 
           label='e1', zorder=1000)

plt.quiver(0, 0, e2[0], e2[1], 
           angles='xy', scale_units='xy', 
           scale=1, color=color_e2, 
           label='e2', zorder=1000)

# 绘制 L1 范数的等距线 
contour = plt.contour(xx, yy, zz,
                      levels=np.arange(1, 9), 
                      cmap='RdYlBu_r')  

# 添加 colorbar
cbar = plt.colorbar(contour)

# 设置坐标轴范围
plt.xlim(-8, 8)
plt.ylim(-8, 8)

# 绘制网格和坐标轴
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xticks(np.arange(-8, 9, 2))
plt.yticks(np.arange(-8, 9, 2))
plt.grid(True, which='both', linestyle='--', linewidth=0.5, color='0.8')

# 设置等比例显示
plt.gca().set_aspect('equal', adjustable='box')
```
